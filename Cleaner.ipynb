```python
import re
import pandas as pd
data = pd.read_csv('train.csv', header=None, skiprows=1, parse_dates=[0,11,12])
print("finished reading train.csv")
data.columns = ["date_time", "site_name", "posa_continent", "user_location_country",
                "user_location_region", "user_location_city", "orig_destination_distance",
                "user_id", "is_mobile", "is_package", "channel", "srch_ci", "srch_co",
                "srch_adults_cnt", "srch_children_cnt", "srch_rm_cnt", "srch_destination_id",
                "srch_destination_type_id", "is_booking", "cnt", "hotel_continent", "hotel_country",
                "hotel_market", "hotel_cluster"]
data.drop(data.index[[406250,1373714,1373715,2374706,3504146,3707429,7780825,778086,8040327,8428261,8428262,
                     15336804,16428469,16695744,17669511,18156670,18575457,31775817,31784369,32146102,32168215,32549132,
                     34815966,34815968,34815969,35412891,36606126,36833435,36833436,37330211]], inplace=True)
```

```python
import datetime
#Finds dates with years beyond 2016
for i in range(0,data.shape[0]):
    if (i % 100000 == 0):
        print(i)
    m1 = re.search(r'(\d+-\d+-\d+)', str(data.iloc[i]['srch_ci']))
    m2 = re.search(r'(\d+-\d+-\d+)', str(data.iloc[i]['srch_co']))
    if(str(data.iloc[i]['srch_ci']) != 'nan'):
        try:
            match = m1.group(0)
            values = match.split('-')
            datetime.datetime(year=int(values[0]),day=int(values[2]),month=int(values[1]))
            if(int(values[0]) > 2016):
                print("bad srch_ci")
                print(data.iloc[i]['srch_ci'])
                print(i)
        except: 
            print("bad srch_ci")
            print(data.iloc[i]['srch_ci'])
            print(i)
    if(str(data.iloc[i]['srch_co']) != 'nan'):
        try:
            match = m2.group(0)
            values = match.split('-')
            datetime.datetime(year=int(values[0]),day=int(values[2]),month=int(values[1]))
            if(int(values[0]) > 2016):
                print("bad srch_co")
                print(data.iloc[i]['srch_co'])
                print(i)
        except: 
            print("bad srch_co")
            print(data.iloc[i]['srch_co'])
            print(i)
    
```

```python
import csv
#script to determine average day and average month
#also determines indices of booking events (to be dropped), which are written to csv
#determines average orig_destination_distance
srch_co_yr = 0.0
srch_co_month = 0.0
num_co_points = 0.0

srch_ci_yr = 0.0
srch_ci_month = 0.0
num_ci_points = 0.0

orig_destination_distance_points = 0
total_distance = 0.0 #sum of non-nan orig_destination_distance values
book_indices = []
for i in range(0,data.shape[0]/2):
    if (i % 100000 == 0):
        print(i)
    if(data.iloc[i]['is_booking'] != 'nan'):
        if(data.iloc[i]['is_booking'] == 1):
            book_indices.append(i)
            
    if(data.iloc[i]['orig_destination_distance'] != 'nan'):
        orig_destination_distance_points += 1
        total_distance += data.iloc[i]['orig_destination_distance']
        
    m1 = re.search(r'(\d+-\d+-\d+)', str(data.iloc[i]['srch_ci']))
    m2 = re.search(r'(\d+-\d+-\d+)', str(data.iloc[i]['srch_co']))
    if(str(data.iloc[i]['srch_ci']) != 'nan'):
            match1 = m1.group(0)
            values1 = match1.split('-')
            year1 = float(values1[0])
            month1 = float(values1[1])
            srch_ci_yr += year1
            srch_ci_month += month1
            num_ci_points += 1 
    if(str(data.iloc[i]['srch_co']) != 'nan'):    
            match2 = m2.group(0)
            values2 = match2.split('-')         
            year2 = float(values2[0])
            month2 = float(values2[1])
            srch_co_yr += year2
            srch_co_month += month2
            num_co_points += 1



wtr = csv.writer(open ('train_book_indices.csv', 'w'), delimiter=',', lineterminator='\n')
for x in book_indices : wtr.writerow (x)

print('Results')
print('total non_nan srch_ci points: ' + str(num_ci_points))
print('total non_nan srch_co points: ' + str(num_co_points))
print('total non_nan orig_destination_distance points: ' + str(orig_destination_distance_points))
print('\n')
print('average srch_ci year' + str(srch_ci_yr/num_ci_points))
print('average srch_ci month' + str(srch_ci_month/num_ci_points))
print('average srch_co year' + str(srch_co_yr/num_co_points))
print('average srch_co month' + str(srch_co_month/num_co_points))
print('average orig_destination_distance' + str(total_distance/orig_destination_distance_points))
```

```python
#script to determine average day and average month
#also determines indices of booking events (to be dropped), which are written to csv
#determines average orig_destination_distance
from datetime import date
import numpy as np
stay_length = 0.0
num_events = 0.0
orig_destination_distance_points = 0.0
total_distance = 0.0 #sum of non-nan orig_destination_distance values
data2 = data.as_matrix(['srch_ci','srch_co','orig_destination_distance'])
for i in range(0,int(data2.shape[0])):
    if (i % 1000000 == 0 and num_events >0):
        print(i)
        print('average stay length: ' + str(stay_length/num_events))
        print('average orig_destination_distance: ' + str(total_distance/orig_destination_distance_points))
            
    if(not np.isnan(data2[i][2])):
        orig_destination_distance_points += 1
        total_distance += (data2[i][2])
        
    m1 = re.search(r'(\d+-\d+-\d+)', str(data2[i][0])) #srch_ci
    m2 = re.search(r'(\d+-\d+-\d+)', str(data2[i][1])) #srch_co
    if(str(data2[i][0]) != 'nan' and str(data2[i][1]) != 'nan'):
            match1 = m1.group(0)
            values1 = match1.split('-')
            year1 = int(values1[0])
            month1 = int(values1[1])
            day1 = int(values1[2])
            
            match2 = m2.group(0)
            values2 = match2.split('-')         
            year2 = int(values2[0])
            month2 = int(values2[1])
            day2 = int(values2[2])
            num_events += 1
            d0 = date(year1, month1, day1)
            d1 = date(year2, month2, day2)
            delta = d1 - d0
            stay_length += delta.days

```
